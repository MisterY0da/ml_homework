```python
import piplite
await piplite.install('seaborn')
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('brainStrokes.csv')
```

```python
print(df)
```

```python
sns.histplot(x='age',data=df,hue='stroke',linewidth=1,multiple='dodge')
plt.title('Влияние возраста на риск инсульта')
plt.show()

# исходя из графика видно, что с увеличением возраста человека увеличивается и риск инсульта
```

```python
sns.countplot(x='stroke',hue='work_type',data=df)
plt.title('Влияние типа работы на риск инсульта')
plt.show()

# большинство людей пострадавших от инсульта были заняты работой в частном секторе
```

```python
sns.countplot(x='stroke',hue='heart_disease',data=df)
plt.title('Влияние сердечных заболеваний на риск инсульта')
plt.show()

# лишь у малой группы людей перенесших инсульт оказались сердечные заболевания
```

```python
sns.countplot(x='stroke',hue='Residence_type',data=df)
plt.title('Влияние места проживания на риск инсульта')
plt.show()

# исходя из графика можно предположить, что городское или сельское окружение не влияет на риск возникновения инсульта
```

```python
sns.countplot(x='stroke',hue='ever_married',data=df)
plt.title('Влияние брака на риск инсульта')
plt.show()

# исходя из графика возможно предположить, что наличие брака вероятно влияет на риск инсульта
```

```python
sns.countplot(x='stroke',hue='gender',data=df)
plt.title('Влияние пола на риск инсульта')
plt.show()

# исходя из графика можно предположить, что пол вероятно не влияет на риск инсульта
```

```python
sns.countplot(x='smoking_status',hue='stroke',data=df)
plt.title('Влияние курения на риск инсульта')
plt.show()

# исходя из графика можно предположить, что курение не сильно влияет на риск инсульта
```

```python
sns.countplot(x='stroke',hue='hypertension',data=df)
plt.title('Влияние гипертонии на риск инсульта')
plt.show()

# исходя из графика наличие шипертонии не сильно влияет ни риск инсульта
```

```python
sns.histplot(x='bmi',data=df,hue='stroke',linewidth=1,multiple='dodge')
plt.title('Влияние индекса массы тела на риск инсульта')
plt.show()

# исходя из графика индекс массы тела влияет на риск инсульта, особенно вблизи значения 30 (первая степень ожирения)
```

```python
sns.histplot(x='avg_glucose_level',data=df,hue='stroke',linewidth=1,multiple='dodge')
plt.title('Влияние среднего уровня глюкозы в крови на риск инсульта')
plt.show()

# исходя из данных, представленных на графике, уровень глюкозы в крови не сильно повлиял на риск инусльта
```
